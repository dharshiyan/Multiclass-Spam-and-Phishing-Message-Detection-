preprocessing step

```python
#preprocessing
def preprocess_text(text):
    text = text.lower()  # lowercase
    
    # remove punctuation
    text = "".join([char for char in text if char not in string.punctuation])
    
    # remove stopwords
    words = text.split()
    words = [word for word in words if word not in stopwords.words('english')]
    
    return " ".join(words)
```

```python
data_1['clean_text'] = data_1['text'].apply(preprocess_text)

print(data_1[['text', 'clean_text']].head())
```

```python
# Remove empty rows
data_1 = data_1[data_1['clean_text'].str.strip() != ""]
```

```python


stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    text = str(text).lower()
    text = "".join(char for char in text if char not in string.punctuation)
    words = text.split()
    words = [word for word in words if word not in stop_words]
    return " ".join(words)

os.makedirs("processed_dataset", exist_ok=True)

chunks = pd.read_csv("processed_dataset/data_5_clean.csv", chunksize=50000)



output_file = "processed_dataset/data_5_fixed.csv"

for i, chunk in enumerate(chunks):
    chunk['text'] = chunk['text'].apply(preprocess_text)
    chunk = chunk[chunk['text'].str.strip() != ""]

    chunk.to_csv(
        output_file,
        index=False,
        mode='a',
        header=(i == 0)   # write header only once
    )

print("Processing complete ✅")
```

```python
import pandas as pd
import os

# Folder where cleaned datasets are stored
folder_path = "processed_dataset"

# Empty list to store dataframes
all_dfs = []

# Loop through all CSV files
for file in os.listdir(folder_path):
    if file.endswith(".csv"):
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path)
        all_dfs.append(df)

# Combine all datasets
combined_df = pd.concat(all_dfs, ignore_index=True)

# Save final dataset
combined_df.to_csv("final_dataset.csv", index=False)

print("All datasets combined successfully ✅")
print("Total rows:", len(combined_df))
```

```python
combined_df = combined_df.sample(frac=1).reset_index(drop=True)
combined_df.to_csv("final_dataset.csv", index=False)
```
